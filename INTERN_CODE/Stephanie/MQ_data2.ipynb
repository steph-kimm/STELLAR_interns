```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
table1 = pd.read_csv('data/Table_1_Periodic.txt')
table1
```

```python
table2 = pd.read_csv('data/Table_2_Non_Periodic.txt')
table2
```

```python
table1 = table1.dropna()
table2 = table2.dropna()
len(table1), len(table2)
```

```python
valid = np.repeat(1,len(table1))
invalid = np.repeat(0,len(table2))
classes = np.append(valid,invalid)
```

```python
tables = pd.concat([table1,table2],axis=0, sort=True)
tables = tables.drop(columns=["Flag","Rper"])

tables
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score
```

```python
X = tables
```

```python
y = classes
```

```python
from sklearn.model_selection import train_test_split

train_X, test_X, train_y, test_y = train_test_split(X, y, random_state = 11)


#NOTICE: below are the same steps as before, except our data sample 
model = DecisionTreeClassifier()
#fit the model to the training set
fit_model = model.fit(train_X,train_y)
#compare predictions from this model to the test set
predicted = fit_model.predict(test_X)
accuracy_score(test_y, predicted)
```

```python
from sklearn import tree
```

```python
tree.plot_tree(model.fit(train_X, train_y)); 
```
