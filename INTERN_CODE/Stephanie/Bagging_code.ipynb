```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
from sklearn.tree import DecisionTreeClassifier
import random
from sklearn import tree
```

```python
#generate fake data

#create 5 empty lists (4 for features 1 for labels)
temps = []
spec_type = []
luminosity = []
mass = []
ages = []

#start a for loop that cycles through range(0,100)
#generate random data
for i in range(0,100):
    temps_range = random.randint(3500,6000)
    spec_type_range = random.randint(0,7)
    luminosity_range = random.randint(1,1000000)##.0001
    mass_range = random.randint(1,100)##1/12
    ages_range = random.randint(100000,13000000)
    
    temps.append(temps_range)
    spec_type.append(spec_type_range)
    luminosity.append(luminosity_range)
    mass.append(mass_range)
    ages.append(ages_range)
    
    
len(temps)
```

```python
d = {'temp': temps, 'spec_type': spec_type,'lum' : luminosity , 'mass': mass }
df = pd.DataFrame(data=d)
df
```

```python
data_table = Table([temps,spec_type,luminosity,mass],names=('temps','spec_type','luminosity','mass'))
```

```python
#bagging selection
random.seed(4)

indices = []
for i in range(0,10):
    indices.append(random.randint(0,100))
    
data_subset = []
labels_subset = []

for i in indices:
    print(i)
    data_subset.append(data_table[i])
    labels_subset.append(data_table[i])
```

```python
#bagging selection
random.seed(2)

indices = []
for i in range(0,10):
    indices.append(random.randint(0,100))
    
data_subset = []
labels_subset = []

for i in indices:
    print(i)
    data_subset.append(data_table[i])
    labels_subset.append(data_table[i])
```

```python
model = DecisionTreeClassifier()
subset_tree_fit = model.fit(data_subset, labels_subset)
```

```python
plt.figure(figsize=(20,15)) 
tree.plot_tree(subset_tree_fit);
```
