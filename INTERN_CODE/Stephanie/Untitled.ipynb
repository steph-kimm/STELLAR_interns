```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
mq_data = pd.read_csv('data/McQuillan_data.csv')
```

```python
mq_data

mass = mq_data['Mass']
rp = mq_data['Prot']
temp = mq_data['Teff']
```

```python
plt.figure(figsize = (15,10))
plt.scatter(temp,rp, s=.1, c=mass)
plt.xlim(6500,3000)
plt.colorbar().set_label(label='Mass', size=20)
plt.xlabel("Temperate(k)", size=25)
plt.ylabel("Rotation Period", size=25)
plt.title("McQuillan 2014 data", size=35)
```
