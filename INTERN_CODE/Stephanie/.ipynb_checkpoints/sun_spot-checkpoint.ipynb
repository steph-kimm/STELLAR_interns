```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('data/sunspots_data.csv')
```

```python
data
```

```python
time = data["linear_year"]
spots = data["daily_sunspots"]
```

```python
plt.figure(figsize = (12,6))
plt.title("Amount of Sun Spots")
plt.plot(time, spots)
plt.xlabel("linear year")
plt.ylabel("daily sun spots")
#plt.xlim(1900,1950)
```
